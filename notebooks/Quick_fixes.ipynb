## Add a first date to the MJO indices

```python
for f in mjofils:
    print(f)
    yrz =int(f.split('.nc')[0].split('_')[-1])
    temp=xr.open_dataset(f)
    print(yrz)
    tt = np.array([cftime.DatetimeNoLeap(yrz, 1,1, 0, 0, 0, 0, 4, 2),cftime.DatetimeNoLeap(yrz, 1,2, 0, 0, 0, 0, 4, 2)])
    dmam=np.nan

    dummy = xr.Dataset(
    {
        "MJO_INDEX": (["time"], [dmam,dmam]),
        "PC1": (["time"], [dmam,dmam]),
        "PC2": (["time"], [dmam,dmam]),
        
    },
    coords={
        "time": tt,
    },
    )
    Fixed = xr.merge([dummy,temp])
    Fixed.to_netcdf(f)
    
```
